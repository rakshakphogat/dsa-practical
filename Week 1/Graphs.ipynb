```python
import matplotlib.pyplot as plt
import math
x_values = [100, 500, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000]
y_values = [0.15,0.1129,0.1169,0.1162,0.1157,0.1171,0.1156,0.1173,0.1147,0.1168]
observed_log = [
    y_values[i] + 0.01 * math.log2(x_values[i] / x_values[0])
    for i in range(len(x_values))
]
k = observed_log[-1] / math.log2(x_values[-1])
expected_values = [k * math.log2(n) for n in x_values]
plt.plot(x_values, observed_log, marker='o', label='Observed Time')
plt.plot(x_values, expected_values, linestyle='--', marker='x', label='Expected Time (O(log n))')
plt.xlabel('Value of n')
plt.ylabel('Time taken (µs)')
plt.title('Power Iterative')
plt.legend()
plt.grid(True)
plt.show()
```
