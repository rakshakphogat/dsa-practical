```python
import matplotlib.pyplot as plt
import math
x_values = [1000, 5000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000]
y_values = [0.3739,0.3482,0.3489,0.3631,0.3579,0.3735,0.3765,0.3946,0.3916,0.384]
observed_log = y_values
k = observed_log[-1] / math.log2(x_values[-1])
expected_values = [k * math.log2(n) for n in x_values]
plt.plot(x_values, observed_log, marker='o', label='Observed Time')
plt.plot(x_values, expected_values, linestyle='--', marker='x', label='Expected Time (O(log n))')
plt.xlabel('Value of n')
plt.ylabel('Time taken (s)')
plt.title('Binary Search')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x_values = [1000, 5000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000]
y_values = [17.5346,30.0307,83.4286,156.951,253.85,367.229,
            504.759,691.056,879.467,1081.83]
observed_quad = y_values
k = observed_quad[-1] / (x_values[-1] ** 2)
expected_values = [k * (n ** 2) for n in x_values]
plt.plot(x_values, observed_quad, marker='o', label='Observed Time')
plt.plot(x_values, expected_values, linestyle='--', marker='x', label='Expected Time (O(n))')
plt.xlabel('Value of n')
plt.ylabel('Time taken (s)')
plt.title('Insertion Sort')
plt.legend()
plt.grid(True)
plt.show()
```
