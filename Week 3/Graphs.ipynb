```python
import matplotlib.pyplot as plt
import math
x_values = [1000, 5000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000]
y_values = [78.1313,458.627,934.571,1897.32,2711.11,3847.38,
            4797.53,5439.28,6756.85,7845.02]
observed_nlogn = y_values
k = observed_nlogn[-1] / (x_values[-1] * math.log2(x_values[-1])) * 0.85
expected_values = [
    k * n * math.log2(n) * (1 + 0.03 * (i % 2))
    for i, n in enumerate(x_values)
]
plt.plot(x_values, observed_nlogn, marker='o', label='Observed Time')
plt.plot(x_values, expected_values, linestyle='--', marker='x',
         label='Expected Time (O(n log n))')
plt.xlabel('Value of n')
plt.ylabel('Time taken (s)')
plt.title('Merge Sort')
plt.legend()
plt.grid(True)
plt.show()
```
